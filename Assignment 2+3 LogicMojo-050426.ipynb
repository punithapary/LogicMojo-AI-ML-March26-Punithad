## 1. Customer Segmentation Engine

You are given a list of customer dictionaries:

```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
```

Task:

1. Write a function segment_customer(customer):
2. If spend divided by income is greater than 0.6 return "High Value"
3. If spend divided by income is between 0.3 and 0.6 return "Medium Value"
4. Otherwise return "Low Value"
5. Create a final dictionary in the format {customer_id: segment}

````python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
def segment_customer(customers):
    result ={}

    for item in customers:
        condition = item["spend"] / item["income"]
        if condition > 0.6:
            result[item["id"]] = "High Value"
        elif 0.3 <= condition <= 0.6:
            result[item["id"]] = "Medium Value"
        else:
            result[item["id"]] = "Low Value"
    return result
final_result = segment_customer(customers)
print(final_result)
    
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
item_dict = {}
def recommend(user_purchases):
    for items in user_purchases.values():
        for item in items:
            if item not in item_dict:
                item_dict[item] = set()
            item_dict[item].update(set(items) - {item})
    return item_dict
recommend = recommend(user_purchases)
print(recommend)

````
