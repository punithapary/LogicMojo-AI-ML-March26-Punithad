Q3. Feature Scaling using NumPy (Very Important)

Context: Most ML models require normalized data.

Dataset: Iris (NumPy array)

Tasks:

1. Apply standardization:

$$
X' = \frac{X - \mu}{\sigma}
$$
	​2. 

Verify:
Mean ≈ 0
Std ≈
on

```python
import seaborn as sns
import numpy as np
titanicDataset = sns.load_dataset("titanic")
numArray = np.array(titanicDataset)
numeric_data = numArray[:, [0, 1]].astype(float)
meanData = np.mean(numeric_data, axis=0)
stdData = np.std(numeric_data, axis=0)
originalData = ( numeric_data - stdData ) / meanData
print("Apply standardization:", originalData)
print("Verify: Mean ≈ 0 Std ≈ on", np.round(np.mean(originalData)), np.round(np.std(originalData)))
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import seaborn as sns
import pandas as pd

dataset6 = sns.load_dataset("titanic")

def age_category(row):
    if row["age"] < 18:
        return "Child"
    elif row["age"] <= 60:
        return "Adult"
    else:
        return "Senior"

dataset6["age_group"] = dataset6.apply(age_category, axis=1)
surData = dataset6.groupby("age_group")["survived"].mean()
print("Compute survival rate per group", surData.head())
surGroup1 = dataset6.groupby("pclass")["survived"].mean()
surGroup2 = dataset6.groupby(["pclass", "sex"])["survived"].mean()
print("Compute survival rate per group", surGroup1, surGroup2)
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np
randomErrors = np.random.normal(loc=0, scale=1, size=1000)
stderror = np.std(randomErrors)
mean_error = np.mean(randomErrors)
print("Compute: Mean Standard deviation errors", np.mean(randomErrors), np.std(randomErrors))
#Identify outliers: Using mean ± 2*std
print("Identify outliers: Using mean ± 2*std", mean_error - 2 * stderror, mean_error + 2 * stderror)
```
