Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:
1. Identify and display all transaction values that appear more than once.
2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
3. Calculate the total spending for the day.
4. Create a new list containing only unique transaction values.
5. Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
#1.Identify and display all transaction values that appear more than once
repeatedValue = [] #to display repeated values
fraudMonitoring = []#to display transactions greated than 5000 values
itemLength = len(transactions)
for i in range(itemLength):
 
    if transactions[i] > 5000 and transactions[i] not in fraudMonitoring:
        fraudMonitoring.append(transactions[i])
        
    countNo = 0
    for j in range(itemLength):
        if transactions[i] == transactions[j]:
            countNo+=1
        if countNo > 1 and transactions[i] not in repeatedValue:
            repeatedValue.append(transactions[i])
print("Display repeated values", repeatedValue)
print("Detect transactions greater than ₹5000 as these may require fraud monitoring.", fraudMonitoring)
print("total spending for the day", sum(transactions))
print("Create a new list containing only unique transaction values.", list(set(transactions)))
print("Sort the unique transaction list in descending order to view the largest transactions first", sorted(transactions, reverse=True))
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:
1. Iterates through the dataset and removes all invalid age values.
2. Stores only valid ages in a new list.
3. Calculates the average age of valid customers.
4. Displays the list of customers whose age is greater than 30.
5. Prints how many invalid entries were removed during the cleaning process.

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
itemLength = len(ages)
validAges = []
realData = []
greaterAge=[]
average = 0
invalidAgeCount=0
for i in range(itemLength):
    if ages[i] is not None and ages[i] not in validAges:
        validAges.append(ages[i])
    if ages[i] is None or ages[i] < 0 :
        invalidAgeCount += 1
print("removes all invalid age values", validAges)
for age in range(len(validAges)):
    if validAges[age] > 0 and validAges[age] not in realData:
        realData.append(validAges[age])
print("Stores only valid ages in a new list", realData)
average = sum(realData) / itemLength
print("Calculates the average age of valid customers", average)
for j in range(len(realData)):
    if validAges[j] > 30 and validAges[j] not in greaterAge:
        greaterAge.append(validAges[j])
print("Displays the list of customers whose age is greater than 30", greaterAge)
print("invalid entries count", invalidAgeCount)
       
```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
Write a Python program that:

1. Separates login hours into even-hour logins and odd-hour logins.
2. Identifies login hours that fall into peak activity time (hours greater than 5).
3. Sorts the login hours in ascending order.
4. Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
loginLength = len(login_hours)
peakData =[]
evenHours =[]
oddhours =[]
peakDataCount=0
for k in range(loginLength):
    if login_hours[k] > 5 and login_hours[k] not in peakData:
        peakData.append(login_hours[k])
    if (login_hours[k] % 2) == 0 and login_hours[k] not in evenHours:
        evenHours.append(login_hours[k])
    if login_hours[k] % 2 != 0 and login_hours[k] not in oddhours:
        oddhours.append(login_hours[k])
print("Separates login hours into even-hour logins and odd-hour logins", evenHours, oddhours )
print("Identifies login hours that fall into peak activity time", peakData)
print("Sorts the login hours in ascending order", sorted(login_hours))
print("Calculates how many logins occurred during peak hours", len(peakData))
```
