```python
import re
data = [

    "User: Mahipal | Phone: 9876543210 | Email: mahi@gmail.com | Amount: Rs.5000",

    "User: Ravi | Phone: 98765abc10 | Email: ravi#gmail.com | Amount: USD300",

    "User: NULL | Phone: 9123456789 | Email: test@yahoo.com | Amount: Rs.7000",

    "User: Ankit | Phone: 9999999999 | Email: ankit@gmail.com | Amount: Rs.0",

    "User: Sita | Phone: 8888888888 | Email: sita@gmail.com | Amount: Rs.-100"

]

convertedData = []
filData = []
errors = []
phoneMatch = re.compile(r'^\d{10}$')
email_pattern = re.compile(r'^[\w\.-]+@[\w\.-]+\.\w{2,6}$')
for record in data:
    fields = record.split(" | ")
    entry = {}

    for item in fields:
        key, value = item.split(": ")
        key = key.lower()

        if key == "user":
            entry["user"] = value.lower()

        elif key == "phone":
            entry["phone"] = value

        elif key == "email":
            entry["email"] = value

        elif key == "amount":
            if value.startswith("Rs."):
                entry["amount"] = int(value.replace("Rs.", ""))
                entry["currency"] = "rs"
            elif value.startswith("USD"):
                entry["amount"] = int(value.replace("USD", ""))
                entry["currency"] = "usd"

    convertedData.append(entry)
    for userData in convertedData: 
        try:
            assert userData["user"] != "null", "User cannot be NULL"
            assert userData["amount"] > 0, "Amount must be greater than 0"
            assert phoneMatch.match(userData["phone"]), "phoneno should be digits"
            assert email_pattern.match(userData["email"]), "email should be in valid format"
            filData.append(userData)
          
        except AssertionError as e:
               print(e)
print(filData)
```
